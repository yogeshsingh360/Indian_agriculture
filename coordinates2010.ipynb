```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

# Load your CSV
df = pd.read_csv('place2001-2010.csv')  # Replace with your actual file path
df = df[['State_Name', 'District_Name']].drop_duplicates()

# Initialize geocoder
geolocator = Nominatim(user_agent="geoapi")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

# Create a location string
df['location'] = df['District_Name'] + ", " + df['State_Name'] + ", India"

# Apply geocoding
df['geo'] = df['location'].apply(geocode)
df['latitude'] = df['geo'].apply(lambda x: x.latitude if x else None)
df['longitude'] = df['geo'].apply(lambda x: x.longitude if x else None)

# Drop helper column if you want
df = df.drop(columns=['geo'])

# Save or view result
df.to_csv('districts_with_coordinates.csv', index=False)
print(df.head())
```
