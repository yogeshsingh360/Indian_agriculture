```python
df['year']=df['year'].astype(int)
```

```python
df_2 = df[df['year'] >= 2010].reset_index(drop=True)
```

```python
df_2= df_2.drop(df_2[df_2["state_name"]=="Andaman And Nicobar Islands"].index)
df_2
```

```python
value=df_2['district_name'].value_counts()
less=value[value<=100].index
less
```

```python
df_2= df_2[~df_2['district_name'].isin(less)]
```
