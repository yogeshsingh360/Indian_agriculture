```python
import pandas as pd 
import numpy as np 
df_1=pd.read_csv('historical_weather_300_data2000.csv')
df_1
```

```python
df_1['date'] = pd.to_datetime(df_1['date'])  # Convert to datetime format

# Add year and month columns
df_1['year'] = df_1['date'].dt.year
df_1['month'] = df_1['date'].dt.month

# Group by state, district, year, and month, and aggregate numerical columns
monthly_df_1 = df_1.groupby(['state', 'district', 'year', 'month']).agg({
    'precipitation_sum': 'mean',
    'relative_humidity_2m_mean': 'mean',  # or 'sum' if needed
    'latitude': 'first',
    'longitude': 'first', 
    'temperature_2m_mean':'mean', 
    'wind_speed_10m_mean':'mean'
}).reset_index()
```

```python
monthly_df_1
```

```python
monthly_df_1.isnull().sum()
```

```python
data=[]
for i in range(0,12):
    if i<10:
        df=pd.read_csv(f'historical_weather_300_data200{i}.csv')
    else:
        df=pd.read_csv(f'historical_weather_300_data20{i}.csv')
    df['date'] = pd.to_datetime(df['date'])  # Convert to datetime format
    
    # Add year and month columns
    df['year'] = df['date'].dt.year
    df['month'] = df['date'].dt.month
    
    # Group by state, district, year, and month, and aggregate numerical columns
    monthly_df = df.groupby(['state', 'district', 'year', 'month']).agg({
        'precipitation_sum': 'mean',
        'relative_humidity_2m_mean': 'mean',  # or 'sum' if needed
        'latitude': 'first',
        'longitude': 'first', 
        'temperature_2m_mean':'mean', 
        'wind_speed_10m_mean':'mean'
    }).reset_index()
    data.append(monthly_df)
    
    
```
